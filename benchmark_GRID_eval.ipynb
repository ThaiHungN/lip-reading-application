```python
# GRID corpus video data
!mkdir -p GRID_data/
# speakers = range(1, 35)
# speakers = range(1, 6)
speakers = [1, 2, 20, 22]
for s in speakers:
  if s != 21: # speaker 21 has no data due to technical oversight
    vid_link = f"https://spandh.dcs.shef.ac.uk//gridcorpus/s{s}/video/s{s}.mpg_vcd.zip"
    vid_filename = f"s{s}.mpg_vcd.zip"
    align_link = f"https://spandh.dcs.shef.ac.uk//gridcorpus/s{s}/align/s{s}.tar"
    align_filename = f"s{s}"
    !curl "$vid_link"
    !wget "$align_link"
    !unzip "$vid_filename" -d GRID_data/video
    !mkdir -p GRID_data/align/"$align_filename"
    !tar xf "$align_filename".tar -C GRID_data/align/"$align_filename"
```
